Цены

```python
prices = pd.read_csv("close_price.csv", parse_dates=True, index_col=0)
prices 
```

Мультипликаторы

```python
pe = pd.read_csv("pe.csv", parse_dates=True, index_col=0)
```

Приводим к единому размеру + заполняем 0 на NaN

```python
prices, volume, pe = pqr.utils.replace_with_nan(prices, volume, pe, to_replace=0)
prices, volume, pe = pqr.utils.align(prices, volume, pe)

print(volume.shape, pe.shape, prices.shape)
```

```python
(prices > 0).sum(axis=1).plot()
```

Пересчитаем объем из шт. акций в доллары

```python
volume = prices * volume * 1000
```

Зададим юниверс

```python
universe = prices > 2
```

```python
returns_calculator = pqr.utils.partial(
    pqr.calculate_returns, 
    universe_returns=pqr.prices_to_returns(prices),
)
```

Бенчмарк

```python
ew_benchmark = pqr.Benchmark.from_universe(
    universe=universe,
    allocator=pqr.equal_weights,
    calculator=returns_calculator,
    name='EW_Benchmark'
)

ew_benchmark.returns
```

#  2 - Шаблоны метрик

Средняя доходность и бенчмарк

```python
table = pqr.metrics.Table()

table.add_metric(
    pqr.utils.partial(
        pqr.metrics.mean_return,
        statistics=True,
        annualizer=12,
    ),
    multiplier=100,
    name="Monthly Mean Return, %",
)

table.add_metric(
    pqr.utils.partial(
        pqr.metrics.alpha,
        benchmark=ew_benchmark,
        statistics=True,
        annualizer=12,
    ),
    multiplier=100,
    name="Alpha, %",
)
```

```python
fig = pqr.metrics.Figure(
    pqr.metrics.compounded_returns,
    name="Returns",
    benchmark=ew_benchmark,
    log_scale=True
)
```

```python
summary = pqr.metrics.Dashboard([table, fig])
summary
```

# Собираем фактор Value из мультипликатора P/E

```python
static_transform = pqr.utils.compose(
    pqr.utils.partial(pqr.factors.filter, universe=universe),
    pqr.utils.partial(pqr.factors.look_back_mean, period=0),
    pqr.utils.partial(pqr.factors.lag, period=1),
    pqr.utils.partial(pqr.factors.hold, period=6),
)
```

```python
value_factor = static_transform(pe)
```

# Собираем портфель

```python
value_portfolios = pqr.factors.backtest_factor_portfolios(
    factor=value_factor, 
    strategies=pqr.factors.split_quantiles(5, better='less'), 
    allocator=pqr.equal_weights,
    calculator=returns_calculator,
    add_wml=False,
)
```

# Смотрим результаты

```python
summary.display(value_portfolios)
```

```python
print(pqr.metrics.alpha(value_portfolios[0], benchmark=ew_benchmark, annualizer=12))
pqr.metrics.trailing_alpha(value_portfolios[0], benchmark=ew_benchmark, annualizer=12).plot()
```
